```python
#1463 백준
x = int(input())

def test(x):
    m = [0 for _ in range(10**6+1)]
    m[2] = 1
    m[3] = 1
    if x > 3:
        for i in range(4,x+1): 
            get = []
            if i%2 == 0 :
                get.append(m[i//2])
            if i%3 == 0 :
                get.append(m[i//3])
            get.append(m[i-1])
            
            m[i] = min(get)+1
    return m[x]

print(test(x))

```

```python
#9095
def test(x) :
    m = [0 for _ in range(11)]
    m[1] = 1
    m[2] = 2
    m[3] = 4
    for i in range(4,x+1):
        m[i] = m[i-1] + m[i-2] + m[i-3]
    return m[x]
    
num = int(input())
for _ in range(num) :
    x= int(input())
    print(test(x))
    
```

```python
#10844 중요
def test() :
    n = int(input())
    dp = [[0 for _ in range(11)] for _ in range(101)]
    for i in range(1,10):
        dp[1][i] = 1
    for i in range(2, n+1):
        dp[i][0] = dp[i-1][1]
        for j in range(1,10):
            dp[i][j] = (dp[i-1][j-1]+dp[i-1][j+1])%1000000000
    sum = 0
    for i in range(10):
        sum += dp[n][i]
    print(sum%1000000000)

test()
    
```

```python
#11057
def test():
    n = int(input())
    dp = [[0 for _ in range(11)] for _ in range(1001)]
    for i in range(10):
        dp[1][i] = 1
    for i in range(2, n+1):
        for j in range(10):
            dp[i][j] = sum(dp[i-1][j:])
            dp[i][j]%=10007
    print(sum(dp[n])%10007)
test()
```

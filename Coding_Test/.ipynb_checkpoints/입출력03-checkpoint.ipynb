```python
#2439
num = int(input())
ans = ""
for i in range(num):
    ans += " "*(num-i-1)
    ans += "*"*(i+1)
    print(ans)
    ans = ""
```

```python
#2445
num = int(input())
for i in range(num-1):
    f = '*'*(i+1)
    s = ' '*(2*(num-i-1))
    t = '*'*(i+1)
    print(f+s+t)
for i in range(num-1,-1,-1):
    f = '*'*(i+1)
    s = ' '*(2*(num-i-1))
    t = '*'*(i+1)
    print(f+s+t)
```

```python
#2522
num = int(input())
for i in range(1,num+1):
    b = ' '*(num-i)
    s = '*'*(i)
    print(b+s)
for i in range(num-1,0,-1):
    b = ' '*(num-i)
    s = '*'*(i)
    print(b+s)
```

```python
#2446
num = int(input())
for i in range(num):
    b = ' '*i
    s = '*'*(2*(num-i)-1)
    print(b+s)
for i in range(num-2,-1,-1):
    b = ' '*i
    s = '*'*(2*(num-i)-1)
    print(b+s)
```

```python
#10991
num = int(input())
for i in range(1,num+1):
    print(' '*(num-i),end='')
    for j in range(i):
        print('*',end=' ')
    print()
```

```python
#1924
x, y = map(int,input().split(' '))
days = -1
if x != 1:
    for i in range(1,x):
        if i in [1,3,5,7,8,10,12]:
            days += 31
        elif i== 2:
            days += 28
        else:
            days += 30
days+= y

ans_list = ['MON','TUE','WED','THU','FRI','SAT','SUN']
print(ans_list[days%7])

```

```python
#10818-max,min 사용 안함
num = int(input())
nl = input().split(' ')
l = []
for i in range(num):
    l.append(int(nl[i]))
b, s = l[0],l[0]
for i in range(1,num):
    if b < l[i]:
        b = l[i]
    if s > l[i]:
        s = l[i]
print(s,b)
```

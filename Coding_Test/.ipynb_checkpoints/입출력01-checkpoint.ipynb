```python
#10950
num = int(input())
k = []
for i in range(num):
    a = input()
    b = a.split(' ')
    k.append(int(b[0]) + int(b[1]))
for l in k :
    print(l)
```

```python
#10951
while True :
    try:
        a,b = map(int, input().split(' '))
        print(a+b)
    except:
        break
```

```python
#10952
while True:
    a,b = map(int, input().split(' '))
    if a== 0 and b == 0 :
        break
    else:
        print(a+b)
```
